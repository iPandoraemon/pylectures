## 例题1: 数值验证问题 

输入多个数(以#结束)，若输入的不是数值，则显示错误，将这些数按整数、浮点数分类显示。

```python
zs = []      # 存放整数的空列表
fds = []     # 存放浮点数的空列表
while True:
    s = input()
    if  s == "#":
        break
    elif s.isdigit():
        zs.append(int(s))
    elif s.count('.')==1 and s.replace('.','').isdigit():
        fds.append(float(s))
    else:
        print("输入错误！")
print("整数:",zs); 
print("浮点数:",fds)
```

## 例题2：显示数字/字母/混合金字塔 

```python
s = input('输入正整数:')
if s.isdigit():
    n = int(s)
    if 0 < n <= 9:
        for i in range(1,n+1):
            a = str(i)*(2*i-1)
            print(' '*(n-i) + a)
    else:
        print('不能超过9行')
else:
    print('不是数字')
```

## 例题3：显示对称金字塔

输入整数n，显示n行对称字母金字塔。





```python
# 第一种产生大写字母表的方法：
zm = "".join([chr(i) for i in range(65,91)])

# 第二种产生大写字母表的方法
# import string
# zm = string.ascii_uppercase

s = input("n=")
if s.isdigit():
    n = int(s)
    for i in range(1,n+1):
        print(" "*(n-i),end="")
        print( zm[:i] + zm[:i-1][::-1])
else:
    print("输入错误")
```

## 例题4：词频统计问题

要求：输入一段英文句子，试统计每个单词出现次数。

句子：Acting as if nothing borne in mind is the best revenge. It's all for myself to live better. Only the best can follow one's inclinations. This is the best state of life. Happiness is about having each tiny wish come ture. Happiness is to find someone who can give you warm and share your life together.

```python
import string                    # 字符串模块

line = input()
for t in string.punctuation:     # 标点符号集字符串
    line = line.replace(t, "")    # 标点替换成空格
wordset = set(line.split())      # 单词集合
 
wordict = {}                     # 初始化用于统计结果的字典
for w in wordset:                # 统计每个单词的频次
    wordict[w] = line.count(w)
    wordlist=sorted(wordict.items()) # 单词排序

for t in wordlist:
    print(t[0]+":"+str(t[1]))
```
