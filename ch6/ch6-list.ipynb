### 内置函数`zip()`

Iterate over several iterables in parallel, producing tuples with an item from each one.

Another way to think of `zip()` is that it turns rows into columns, and columns into rows. This is similar to transposing a matrix.

`zip()` is lazy: The elements won’t be processed until the iterable is iterated on, e.g. by a for loop or by wrapping in a list.


```python
l0 = [1, 2, 3, 4]
l1 = ['a', 'b', 'c']
l2 = [5, 6]
l3 = zip(l0, l1, l2)
l3
```

```python
list(l3)
```

## 列表推导式

```python
l0 = [x*2 for x in range(5)]
l0
```

```python
l1 = [x**2 for x in range(5)]
l1
```

```python
l2 = [x*y for x in range(10) if x%2 == 0 for y in range(10) if y%3 == 0]
l2

l2 = []
for x in range(10):
    if x%2 == 0:
        for y in range(10):
            if y%3 == 0:
                l2.append(x*y)
```

解释：

* x=0, y=0 --> x\*y=0
*x=0, y=1
* x=0, y=2
* x=0, y=3 --> x\*y=0
* ...
* x=0, y=6 --> x\*y=0
* ...
* x=0, y=9 --> x\*y=0
* x=1, y=0
* ...
* x=1, y=9
* x=2, y=0 --> x\*y=0
* x=2, y=1
* ...
* x=2, y=3 --> x\*y=6
* x=2, y=4
* x=2, y=5
* x=2, y=6 --> x\*y=12
* ...
* x=2, y=9 --> x\*y=18
* x=3, y=0
* ...
* x=3, y=9
* x=4, y=0 --> x\*y=0
* x=4, y=1
* ...
* x=4, y=3 --> x\*y=12
* ...
* 以此类推

```python
l3 = [k*k for k in range(1, 21) if k%3==0 or k%5==0]
l3
```

```python
# 等价于：
l4 = []
for k in range(1, 21):
    if k%3 == 0 or k%5 == 0:
        l4.append(k*k)
l4
```
